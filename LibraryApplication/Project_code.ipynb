# `Book` Class
A Base class representing a book with the below parameters-

### Parameters:
- `date`: Date on which a particular book is added to the catalog.
- `title`: Title of the book.
- `genre`: Book’s genre.
- `type`: Type of the book.
- `rating`: Customers’s rating for the book.
- `price`: Price of the book.
- `average reading time`: Average time required to read the book.


```python
class Book:
  def __init__(self, date, title, genre, type, rating, price, average_reading_time):
    self.date = date
    self.title = title
    self.genre = genre
    self.type = type
    self.rating = rating
    self.price = price
    self.average_reading_time = average_reading_time
    self.unique_id=id(self)

  def __str__(self):
    return f'{self.unique_id},{self.date},{self.title},{self.genre},{self.type},{self.rating},{self.price},{self.average_reading_time}'
```

-   Generating the dataframe 

```python
book_data=pd.read_csv('data.csv')
book_data
```

-   Studying the Null Values present in the data frame

```python
null_values_by_title = book_data.groupby('title').apply(lambda x: x.isna().sum())
print(null_values_by_title)
```

-   Imputing the missing values of Average Reading time by grouping the datset beased on titles and imputing the Mean of the reading times against the null values
-   Imputing the missing values of Ratings by grouping the dataset based on titles and imputing the mode of the rating against the null values

```python
#filling the average reading time for each title for the missing value
book_data['average_reading_time']=book_data.groupby('title')['average_reading_time'].transform(lambda x:x.fillna(x.mean()))

# Filling missing 'rating' with the mode of the same title; using the first mode
book_data['rating'] = book_data.groupby('title')['rating'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x))

```

-   For filling in the  missing dates we sorted the data set wrt to time and opted for filling in the previous date of sale into the subsequent date row 

As  pricing is more dynamic we opted for creating weekly data frames for each title and calculating mean for that particular week to impute the missing price value for that title

```python
#sorting the entire data frame by date value
book_data['date'] = pd.to_datetime(book_data['date'], errors='coerce')
data_sorted_by_date = book_data.sort_values('date')
data_sorted_by_date.head()
```

-   Created separate dataframes for each book 

```python
# Create a dictionary of dataframes, each containing data for a specific book title
dataframes_by_title = {title: book_data[book_data['title'] == title] for title in book_data['title'].unique()}
# Display the keys to confirm creation of separate dataframes for each title
dataframes_by_title.keys()
```

```python
#different dictionaries created
tdoyg_df=dataframes_by_title['The Diary of a Young Girl']
sj_df=dataframes_by_title['Steve Jobs']
gg_df=dataframes_by_title['Gone Girl']
motoe_df=dataframes_by_title['Murder on the Orient Express']
abhot_df=dataframes_by_title['A Brief History of Time']
mx_df=dataframes_by_title['Malcolm X']
s_df=dataframes_by_title['Sapiens']
tsg_df=dataframes_by_title['The Selfish Gene']
motoe_df=dataframes_by_title['Murder on the Orient Express']
```

Imputing the missing price with the previous available after we have segregated the data frames.As doing this at the earlier stage of would have led to generation erroneous data

```python
#imputing the previous rows data
tdoyg_df['date'].fillna(method='ffill', inplace=True)
tdoyg_df['price'].fillna(method='ffill', inplace=True)
```

Creating weekly data frames to generate mean price for each week and imputing that into the missing price points

```python
#Week wise dataframe creation to impute 0 value for pricing for The diary of a Young Girl
# Set the date column as the index
tdoyg_df.set_index('date', inplace=True)
# Create a dictionary to store each week's dataframe
weekly_dataframes = {f"Week {idx + 1}": week_df for idx, (week, week_df) in enumerate(tdoyg_df.resample('W'))}
```

```python
for week_label, df in weekly_dataframes.items():
    average_price = df['price'].mean()  # Just an example operation
    print(f"{week_label} - Average Price: {average_price}")
```

```python
# Loop through each week's dataframe in the weekly dataframes dictionary
for week_label, df in weekly_dataframes.items():
    # Calculate mean of 'price' excluding zero values
    mean_price = df[df['price'] != 0.00]['price'].mean()

    # Replace zeros in the 'price' column with the mean price
    df['price'] = df['price'].replace(0.00, mean_price)

```

```python

# Concatenate all weekly dataframes stored in the dictionary back into one dataframe
combined_df = pd.concat(weekly_dataframes.values())
# Reset the index
combined_df.reset_index(inplace=True)
# Assign the result back
tdoyg_df = combined_df
# Display the combined dataframe to verify
print(tdoyg_df)


```

```python
#Time Series plot depicitng Weekly sale of the books
book_df.set_index('date', inplace=True)

# Create a dictionary to store each week's dataframe
weekly_dataframes = {f"Week {idx + 1}": week_df for idx, (week, week_df) in enumerate(book_df.resample('W'))}

# Initialize lists to store the weekly sums and week labels for plotting
week_sums = []
week_labels = []

# Loop over all weeks stored in the weekly_dataframes dictionary
for week_label, df in weekly_dataframes.items():
    if df.empty:
        continue  # Skip if the dataframe for a week is empty
    # Calculate the total price for the week
    total_price = df['price'].sum()
    # Append the results for plotting
    week_sums.append(total_price)
    week_labels.append(df.index[0])  # Use the first date of each week as the label for the x-axis

# Plotting the weekly total prices
plt.plot(week_labels, week_sums, marker='o', linestyle='-', color='blue')
plt.title('Time Series Plot of Sales for Each Week Throughout the Year')
plt.xlabel('Week Starting Date')
plt.ylabel('Total Sales each week')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate date labels for better readability
plt.tight_layout()  # Adjust layout to make room for rotated labels
plt.show()

```

This time series plot shows the total sales for each week throughout the year. The sales start at a very high level at the beginning of the year, then sales decline steadily until reaching a low point. The overall trend shows a decline in sales from the beginning to the end of the year.

```python
# Initialize lists to store the weekly means and week labels for plotting
week_means = []
week_labels = []

# Loop over all weeks stored in the weekly_dataframes dictionary
for week_label, df in weekly_dataframes.items():
    if df.empty:
        continue  # Skip if the dataframe for a week is empty
    # Calculate the mean price for the week
    mean_rating = df['rating'].mean()
    # Append the results for plotting
    week_means.append(mean_rating)
    week_labels.append(df.index[0])  # Use the first date of each week as the label for the x-axis

# Plotting the weekly mean prices
plt.plot(week_labels, week_means, marker='o', linestyle='-', color='blue')
plt.title('Time Series Plot of ratings for Each Week Throughout the Year')
plt.xlabel('Week Starting Date')
plt.ylabel('Mean rating')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate date labels for better readability
plt.tight_layout()  # Adjust layout to make room for rotated labels
plt.show()
```

This time series plot shows the ratings for each week throughout the year.
